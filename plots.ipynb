Sinusoid (10-Shot)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dummy data
gradient_steps = np.arange(10)
maml_error = [3.099134, 0.39957926, 0.22667727, 0.17168239, 0.14971985, 0.13533683, 0.12700392, 0.12001339, 0.1161437, 0.11265253]
pretrained_error = [3.0803912, 2.883526, 2.778726, 2.6977768, 2.6307385, 2.5727434, 2.520922, 2.4749553, 2.3953836, 2.3606632]
oracle_error = np.full_like(gradient_steps, 0)  # Constant low error for oracle

# Plotting
plt.figure(figsize=(7, 4))
plt.plot(gradient_steps, maml_error, 'g-o', label="MAML")
plt.plot(gradient_steps, pretrained_error, 'b--s', label="pretrained")
plt.plot(gradient_steps, oracle_error, 'r-.v', label="oracle")

# Labels and Title
plt.xlabel("number of gradient steps", fontsize=12)
plt.ylabel("mean squared error", fontsize=12)
plt.title("k-shot regression for sinusoid, k=10", fontsize=14)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

# Show the plot
plt.show()
```
